```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Bad plots, wrong data and correct them

```python

x = ["Now","JAN-1-2013"]
y = [35, 39.6]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

imagen = mpimg.imread('graph1.jpg')
axs[0].imshow(imagen)
axs[0].axis('off')

colors = ["#FFCC33"]
axs[1].bar(x, y, alpha=1, color=colors, edgecolor="gray", linewidth=2)
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Porcentage (%)')
axs[1].set_title('TOP TAX RATE')
axs[1].grid(True, linewidth=0.5, alpha=0.5, linestyle='--')
axs[1].set_ylim([0,45])
# Aade el valor del porcentaje dentro de cada barra
for i, v in enumerate(y):
    axs[1].text(i, v+1, str(v)+'%', color='black', ha='center')
```
